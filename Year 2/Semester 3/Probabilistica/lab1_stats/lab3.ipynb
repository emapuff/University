```python
import random
from random import choices, sample
from math import comb, perm

def probabilitate(simulari):
    a = 0 
    b = 0
    for i in range(simulari):
        rezultat = sample(['R','B','G'],counts=[5,3,2],k=3)
        if 'R' in rezultat:
            a +=1
            if rezultat == ['R','R','R']:
                 b+=1
            
    return b/a 
    
print(probabilitate(1000))
```

```python
from random import randrange
from matplotlib.pyplot import bar, hist, grid, show, legend
data = [randrange(1,7) for _ in range(500)]
bin_edges = [k+0.5 for k in range(7)]
hist(data, bin_edges, density = True, rwidth = 0.9, color = 'yellow', edgecolor = 'black',
alpha = 0.5, label = 'frecvente relative')
distribution = dict([(i,1/6) for i in range(1,7)])
bar(distribution.keys(), distribution.values(), width = 0.85, color = 'violet', edgecolor = 'black',
    alpha= 0.6, label = 'probabilitati')
legend(loc = 'lower left')
grid()
show()
```
