```python
from scipy.stats import norm
from numpy import mean, std, linspace
from matplotlib.pyplot import show,hist, grid,legend,xticks,plot
# sigma iti spune cat de variate sunt valorile pe care le obti 
# cat trebuie sa te duci in stranga si in dreapta ca sa prinzi valorile posibile 
#cu cat e mai mare sima cu atat valorile sunt mai multe 
#tip problema 
data = norm.rvs(loc=165,scale = 10, size = 1000)
hist(data, bins = 14, density = True, range=(130,200),color = 'teal', label='Frecvente relative',)
x = linspace(130,200,1000)
plot(x,norm.pdf(x,loc=165,scale=10),'m-', label='Functie densitate')
xticks(range(130,205,5))
legend(loc = 'upper right')
grid()
show()
```

```python
print(data.mean())
print("sigma -", data.std())
print("cate pers - ",len(data))
print("media -", sum((160< x < 170 for x in data))/len(data))

print(mean((160<=data)&(data<=170)))
print(norm.cdf(170,loc=165,scale = 10)-norm.cdf(160,loc=165,scale=10))
```

```python
from scipy.stats import norm,expon,uniform,bernoulli
from numpy import mean, std, linspace
from matplotlib.pyplot import show,hist, grid,legend,xticks,plot

sample = [imp * expon.rvs(loc=0, scale =5) + (1-imp)*uniform.rvs(5) for imp in bernoulli.rvs(p=0.4, size = 1000)]
medie = mean (sample)
print(medie)
print(std(sample))
print(expon.cdf(5,scale=5)*0.4+uniform.cdf(5,loc=4,scale=2)*0.6)
```

```python
from scipy.stats import uniform
from numpy import exp, mean
from scipy.integrate import quad

n=5000
u = uniform.rvs(loc = -1, scale = 4 , size = n)
g = lambda x: exp(-x**2)
mean(4*g(u)) #inmultim cu 3-(-1)
print(quad(g,-1,3))
#valoare si eroare 
```
