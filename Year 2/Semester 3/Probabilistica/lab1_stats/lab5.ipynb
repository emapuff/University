```python
# Problema 2
# a).
from scipy.stats import uniform
from math import log

def genereaza_exponentiala(alpha, N):
    if alpha <= 0:
        raise ValueError("Parametrul alpha trebuie să fie strict pozitiv.")

    U = uniform.rvs(size=N)
    X = [-log(1 - u) / alpha for u in U]
    return X

# alpha = 2.0
# N = 1000

# numere_exponentiale = genereaza_exponentiala(alpha, N)

# Aplicatie
from matplotlib.pyplot import show, hist, grid, legend, xticks, plot
from scipy.stats import expon
from random import seed

alpha = 1 / 12
scale = 1 / alpha
N =1000

data = genereaza_exponentiala(alpha, N)
hist(data, bins=12, density=True, range=(0, 60), color='blue', alpha=0.6, edgecolor='black', label='Simulare')
x = range(60)
plot(x, expon.pdf(x, loc=0, scale=scale), 'r-', label='Densitate teoretică')

xticks(range(0, 61, 5))
grid()
legend()
show()

#b).
print(sum([x > 5 for x in data])/len(data))

#Probabilitatea teoretica
print(1 - expon.cdf(5, loc=0, scale=1/alpha))
```
