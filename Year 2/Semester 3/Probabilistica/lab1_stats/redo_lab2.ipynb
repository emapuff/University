```python
import random


#prin simulari repetate probabilitatea ca intr un  grups sa fie mai multi cu aceiasi zi de nastere\
from random import randint

def generate_bday(n): #generam mai multe zile 
    bday = []
    for _ in range(1,n):
        bday.append(randint(1,365))
    return bday

def simulate(n,nr):
    c =0 
    for _ in range(n):
        bday = generate_bday(nr)
        if len(bday) > len(set(bday)):
            c = c+ 1
    return c/n #returnam cazuri favorabile/cazuri posibile


    
```

```python
# desene
from matplotlib.pyplot import plot, grid, title, show
from matplotlib.pyplot import plot, grid, title, show
xs = range(2,51)
ys = [simulate(1000,x) for x in xs]
plot(xs,ys,'r*')
grid()
show()
```
