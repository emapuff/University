```python
from random import randint

#1.a

def generateBirthday(n):
    birthday = []
    for i in range(0,n):
        birthday.append(randint(1,365))
    return birthday

def duplicity(birthday):
    set1 = set(birthday)
    unique = list(set1)
    return len(unique)

def function(n):
    birthday = generateBirthday(n)
    
print(generateBirthday(26))
print(duplicity(generateBirthday(26)))
```

```python
def simulare(n,nr):
    cp = 0
    for i in range(0,n):
        list = generateBirthday(nr)
        #print(list)
        if len(list)-duplicity(list)>0:
            cp = cp+1
    return cp/n

print(simulare(1000,26)*100,"%")
```

```python
from matplotlib.pyplot import plot, grid, title, show
xs = range(2,51)
ys = [simulare(1000,x) for x in xs]
plot(xs,ys,'r*')
grid()
show()
```

```python
from matplotlib.pyplot import axis, plot, show
from random import random
from math import dist
axis('square')
axis((0, 1, 0, 1))
contor = 0
for i in range(0,20000):   
    F=[random(),random()]
    if dist(F,(0.5,0.5))<0.5:
        contor+=1
        plot(F[0],F[1],'yo')
    else:
         plot(F[0],F[1],'ro')
show()
print(contor/20000*4)
```
