```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from math import log10

dataFrame = pd.read_csv("surveyDataSience.csv")

experience_mapping = {
    "I have never written code": 0,
    "< 1 year": 0.5,
    "1-3 years": 1.5,
    "3-5 years": 4,
    "5-10 years": 7.5,
    "10-20 years": 15,
    "20+ years": 25
}

experience_data = dataFrame["Q6"].map(experience_mapping)
experience_data.dropna(inplace=True)


plt.hist(experience_data, bins=np.logspace(0, np.log10(25), 6), edgecolor="black")
plt.xscale("log")  
plt.xlabel("Ani de experien")
plt.ylabel("Frecven")
plt.xticks([1, 3, 5, 10, 20], labels=["1", "3", "5", "10", "20+"])  
plt.title("Distribuia experienei n programare (original)")
plt.tight_layout()
plt.show()


experience_log = [log10(p) for p in experience_data if p > 0]


bins_log = np.linspace(min(experience_log), max(experience_log), 6)

plt.hist(experience_log, bins=bins_log, edgecolor="black",color="yellow")
plt.xlabel("Ani de experien (log10)")
plt.ylabel("Frecven")
plt.xticks(bins_log, labels=[f"{10**x:.1f}" for x in bins_log])  
plt.title("Distribuia experienei n programare (log10 normalizat)")
plt.tight_layout()
plt.show()
```

<b> problema 2 - valorile pixelilor din imagini </b>

```python

from PIL import Image
import os
import random
images_list = os.listdir("images")
image = Image.open(os.path.join("images", images_list[random.choice(range(len(images_list)))]))
image.show()
pixels = np.array(image)

plt.hist(pixels.flatten(), bins=256, color='red', edgecolor='black')
plt.title("Distribuia valorilor pixelilor")
plt.xlabel("Intensitate pixel")
plt.ylabel("Frecven")
plt.show()

pixels_minmax = (pixels - pixels.min()) / (pixels.max() - pixels.min())
plt.subplot(1, 2, 2)
plt.hist(pixels_minmax.flatten(), bins=256, color='gray', edgecolor='black')
plt.title("Min-Max Normalization")
plt.xlabel("Intensitate pixel")
plt.ylabel("Frecven")

plt.tight_layout()
plt.show()
```

<b> problema 3 </b> - numarul de aparitii a cuvintelor la nivelul unei propozitii.

```python
import re
from nltk.probability import FreqDist
with open("texts.txt", 'r', encoding='utf-8') as f: text = f.read()
    
sentences = re.split(r'[.!?]', text)
sentences = [p.strip() for p in sentences if p.strip()]

for sentence in sentences:
    words = re.findall(r'\b\w+\b', sentence)
    cleaned_words = [word for word in words if word.isalnum() and not word.isdigit()]
    word_freq = FreqDist(cleaned_words)
    total_words = len(cleaned_words)
    normalization_factor = np.sqrt(sum(frequency ** 2 for frequency in word_freq.values()))
    word_list = []
    normalized_freq_list = []

    for word, frequency in word_freq.items():
        if normalization_factor != 0:  
            normalized_frequency = frequency / normalization_factor * 100
        else:
            normalized_frequency = 0  
        
        word_list.append(word)
        normalized_freq_list.append(normalized_frequency)
        print(f'{word}: {frequency}/{total_words} ({round(normalized_frequency, 2)}%)')

    plt.figure(figsize=(18, 10))
    plt.bar(word_list, normalized_freq_list)
    plt.title(f'Distribuia frecvenelor cuvintelor (Normalizare pe baza normei vectoriale) - Propoziia: {sentence}')
    plt.xlabel('Cuvnt')
    plt.ylabel('Frecven normalizat')
    plt.xticks(rotation=90)
    plt.show()
    
    
    
```
